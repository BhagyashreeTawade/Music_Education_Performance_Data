1.How many male and female students are there ?

```python

gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightblue', 'pink'], figsize=(6, 6))
plt.title('Gender Distribution')
plt.ylabel('')  # Removes default y-axis label
plt.show()
```

2.what factors most influence the performance score ?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Performance_Score', data=df)
plt.title('Age vs Performance Score')
plt.xlabel('Age')
plt.ylabel('Performance Score')
plt.show()
```

3.difference in performance score between male and female

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Gender",y="Performance_Score",data=df)
plt.title("Performance Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Performance Score")
plt.show()
```

4.average engagement score for students across diff class level

```python
average_engegment = df.groupby("Class_Level")["Engagement_Score"].mean()
average_engegment.plot(kind="bar",figsize=(8,6),color="Red")

plt.title('Average Engagement Score by Class Level')
plt.xlabel('Class Level')
plt.ylabel('Average Engagement Score')
plt.show()
```

5.average accuracy by lesson type

```python
avg_accuracy_by_lesson_type = df.groupby('Lesson_Type')['Accuracy'].mean()
avg_accuracy_by_lesson_type.plot(kind="bar",figsize=(8,6),color="lightgreen")
plt.title("Average Accuracy by Lesson Type")
plt.xlabel("Lesson Type")
plt.ylabel("Accuracy")
plt.show()
```

6.How do students blood pressure level changed across diff lesoon type ?

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Lesson_Type',y='Blood_Pressure',data=df)
plt.title("Blood Pressure by Lesson Type")
plt.xlabel("Lesson Type")
plt.ylabel("Blood_Pressure")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Class_Level',y='Performance_Score',data=df)
plt.title("Performance score by class level")
plt.xlabel("Class level")
plt.ylabel("Performance Score")
plt.show()
```

```python
#8.Average stress and Average Engagement score
avg_stress_level = df["Stress_Level"].mean()
avg_engagement_score = df["Engagement_Score"].mean()

data = {
    "Category": ["Stress_Level", "Engagement_Score"],
    "Average": [avg_stress_level, avg_engagement_score]
}
chart_df = pd.DataFrame(data)


plt.bar(chart_df["Category"],chart_df["Average"],color=["blue","green"])
plt.figure(figsize=(8,6))
plt.title("Average Stress Level and Average Engagement Score")
plt.xlabel("Category")
plt.ylabel("Average")
plt.show()
```

8.Average stress and Average Engagement score

```python
avg_stress_level = df["Stress_Level"].mean()
avg_engagement_score = df["Engagement_Score"].mean()

data = {
    "Category": ["Stress_Level", "Engagement_Score"],
    "Average": [avg_stress_level, avg_engagement_score]
}
chart_df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))  
plt.bar(chart_df["Category"], chart_df["Average"], color=["blue", "green"])
plt.title("Average Stress Level and Average Engagement Score") 
plt.xlabel("Category")  
plt.ylabel("Average") 
plt.show()
```

9.Distribution of Instrument type by Gender

```python
#group by Instrument_Type and Gender, and count the occurrences
grouped_data = df.groupby(['Instrument_Type', 'Gender']).size().unstack()

grouped_data.plot(kind="bar",figsize=(8,6),colormap="viridis",edgecolor="black")
plt.title("Distribution of instrument type by Gender")
plt.xlabel("Instrument Type")
plt.ylabel('Count',)
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

10.What is the age distribution of students?

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=8, color='lightgreen', edgecolor='black')
plt.title('Age Distribution of Students', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

11.What is the tempo distribution?


```python

sns.kdeplot(df['Tempo'], fill=True, color='purple')
plt.title('Tempo Distribution', fontsize=14)
plt.xlabel('Tempo', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```
